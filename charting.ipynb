```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x = [1, 2, 3]
y = [43, 54, 61]

x2 = [9, 2, 1]
y2 = [23, 12, 56]

plt.plot(x, y, label='First', linewidth=4)
plt.plot(x2, y2, label='Second')

plt.legend()
plt.show()
```
